```python
import pandas as pd
import sqlite3
#імпортує пандас для роботи з таблицями та бд sqlite3
pd.set_option('display.max_columns', 100)
pd.set_option('display.expand_frame_repr', False)
#налаштовую вигляд таблиці
df1 = pd.read_csv("/Users/admin/Downloads/itunes_dataset/20190210.txt" ,delimiter="\t")
#Зчитую кожен файл, виконавши цю команду він автоматично вигрузиться в бд
df1.drop_duplicates()
#використовуючи контекстний менеджер, вигружаю інформацію в базу даних
with sqlite3.connect("betterMe.db") as conn:
        df1.to_sql('table', conn, if_exists='append', index=False)
```
